```python
df=pd.read_csv("ufc-fighters-statistics.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['nickname'],inplace=True) #nickname not necessary
df.head()
```

```python
df=df.rename(columns={'height_cm':'height','weight_in_kg':'weight','reach_in_cm':'reach','date_of_birth':'dob'})
numcol=['height','weight','reach']
for i in numcol:
  df[i]=df[i].fillna(df[i].median()) #filling all numeric null values with median
df.head()
```

```python
for i in ['stance']:
  df[i]=df[i].fillna('Unknown') #filling all stance null values with "Unknown"
df['dob']=pd.to_datetime(df['dob']) #converting dob column to date time format
df.head()
```

```python
df['win_rate']=df['wins']/(df['wins']+df['losses']+df['draws'])
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True))
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['height'],bins=20,edgecolor='black')
plt.title('Heights histogram')
plt.xlabel('heights in cm')
plt.ylabel('Number of fighters')
plt.show() #most fighters haave height between 170-185
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['weight'],bins=20,edgecolor='black')
plt.title('Weights histogram')
plt.xlabel('Weights in kg')
plt.ylabel('Number of fighters')
plt.show()
```

```python
df[df['weight']>300]
```

```python
df.loc[df['weight']>300, 'weight']=df['height'].quantile(0.75)+1.5*(df['height'].quantile(0.75)-df['height'].quantile(0.25))
df.loc[4069,'weight']
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['weight'],bins=20,edgecolor='black')
plt.title('Weights histogram')
plt.xlabel('Weights in kg')
plt.ylabel('Number of fighters')
plt.show() #majority fighters have weight between 60-90
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['height'], df['weight'], alpha=0.8)
plt.title('Height vs Weight')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['win_rate'], bins=20, edgecolor='black')
plt.title('Win Rate')
plt.xlabel('Win Rate')
plt.ylabel('Fighters')
plt.show()
```
